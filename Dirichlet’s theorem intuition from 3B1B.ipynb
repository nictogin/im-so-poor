```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def EratoSieve(n):
    prime = [True for i in range(n + 1)]
    numbers = list()
    p = 2
    while (p * p <= n):
        if (prime[p] == True):
            for i in range(p * 2, n + 1, p):
                prime[i] = False
        p += 1
    prime[0]= False
    prime[1]= False
    for p in range(n + 1):
        if prime[p]:
            numbers.append(p)
    return numbers
```

```python
Polar = EratoSieve(10000)
plt.polar(Polar,Polar,'o')
plt.show()
```
