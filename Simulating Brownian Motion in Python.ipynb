```python
import numpy as np
import matplotlib.pyplot as plt

def main():
    n = 10000
    T = 1.
    time = np.linspace(0.,T,n)
    dt = time[1]-time[0] #time steps
    #Bt2-Bt1 ~ Normal with mean 0 and variance t2-t1 (increments of brownian motion)
    dB = np.sqrt(dt)*np.random.normal(size=(n-1)) #default mean 0 variance 1
    B0 = np.zeros(shape=(1,))  #we use np to generalize it later
    B = np.concatenate((B0,np.cumsum(dB)))
    plt.plot(time,B)
    plt.show()
    
if __name__ == '__main__':
    main()
```

```python

import matplotlib.pyplot as plt
import numpy as np


def quadratic_variation(B):
    return np.cumsum(np.power(np.diff(B, axis=0, prepend=0.), 2), axis=0) 


def main():
    n = 10000
    d = 10 #number of brownian motions we wanna simulate
    T = 1.
    times = np.linspace(0., T, n)
    dt = times[1] - times[0]
    # Bt2 - Bt1 ~ Normal with mean 0 and variance t2-t1
    dB = np.sqrt(dt) * np.random.normal(size=(n - 1, d)) #(time,space/index of BM)
    B0 = np.zeros(shape=(1, d)) #starts at zero
    B = np.concatenate((B0, np.cumsum(dB, axis=0)), axis=0)     ##cumsum in time axis but not dimension/space, same as the concatenation 
    plt.plot(times, B)
    plt.plot(times, quadratic_variation(B))#quadratic variation of brownian motion is t at time t(sum of square of differences)
    plt.show()


if __name__ == '__main__':
    main()
    
#for more advanced stat theory, the law of iterated logorithms
```
