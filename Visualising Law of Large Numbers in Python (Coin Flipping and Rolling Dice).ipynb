```python
import matplotlib.pyplot as plt
import random 

heads_tails = [0,0]

for n in range(10000):
    heads_tails[random.randint(0,1)] += 1
    #if n % 100 == 0:
        #plt.bar([0,1], heads_tails, color=('Blue','Red'))
        #plt.pause(0.05)
plt.bar([0,1], heads_tails, color=('Blue','Red'))
plt.show()
```
